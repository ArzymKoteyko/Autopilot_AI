```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import trange
import yfinance as yf 
```

```python
MMM              = yf.Ticker('MMM')
American_Express = yf.Ticker('AXP')
Apple            = yf.Ticker('AAPL')
Boeing           = yf.Ticker('BA')
Caterpillar      = yf.Ticker('CAT')

DJI              = yf.Ticker('DJI')
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (40,20))

sns.set(style='whitegrid')

ax1.plot(MMM.history(period='5y')['Close'], label='MMM')
ax1.plot(American_Express.history(period='5y')['Close'], label='American_Express')
ax1.plot(Apple.history(period='5y')['Close'], label = 'Apple')
ax1.plot(Boeing.history(period='5y')['Close'], label = 'Boeing')
ax1.plot(Caterpillar.history(period='5y')['Close'], label = 'Caterpillar')
ax1.legend(fontsize = 'xx-large')

ax2.plot(DJI.history(period='10y')['Close'], label = 'DJI')
ax2.legend()
```

```python
Ubisoft = yf.Ticker('UBI.PA')
Blizzard_Activision = yf.Ticker('ATVI')

time_fraction = len(Ubisoft.history(period='5y')) / 100
df = Ubisoft.history(period='5y')
df.insert(0, 'Tf', [x // time_fraction for x in range(len(Ubisoft.history(period='5y')))])

sns.set(style='whitegrid', palette=sns.color_palette("RdBu", n_colors=2))
plt.figure(figsize = (40,20))
sns.lineplot(y = 'Close', x = 'Tf', data = df, label = 'Ubisoft')

time_fraction = len(Blizzard_Activision.history(period='5y')) / 100
df = Blizzard_Activision.history(period='5y')
df.insert(0, 'Tf', [x // time_fraction for x in range(len(Blizzard_Activision.history(period='5y')))])

sns.lineplot(y = 'Close', x = 'Tf', data = df, label = 'Blizzard_Activision')
plt.legend(fontsize = 'xx-large')
```

```python
A = [4,5,6,2,3,7,14,161,2,3,7]

s = 0
for k in range(1,10):
    if A[k-1]<A[k]<A[k+1]:
        A[k] = 2 * A[k]
        s = s + A[k-1]
    
print(s)
```

```python
from random import randint

def foo(A):
    minim = 15001
    for i in range(len(A)):
        if A[i] < minim and A[i] % 3 != 0:
            minim = A[i]
    print(minim)
    for i in range(len(A)):
        if A[i] % 3 != 0:
            A[i] -= minim
    return A
A = [2,3,5,9,12,4] #   [0, 3, 3, 9, 12, 2]
print(A, foo(A))
A = [6,6,7,9,12,4] #   [6, 6, 3, 9, 12, 0]
print(A, foo(A))
```
